##Numpy Array Operations

**1. Create a 1D NumPy Array**


```python
# Create a Numpy

import numpy as np
temparature_w1 = np.array([22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9])

print(temparature_w1)
```

**2. Inspection and Properties**

```python
# Inspect the shape, data type, and the number of elements in the array.Inspect the shape of the temperatures_w1 array.

# Shape
print("Shape:", temparature_w1.shape)
```

```python
# Data type
print("Data type:", temparature_w1.dtype)
```

```python
# Size/Number of elements
print("Size:", temparature_w1.size)
```

**3. Array Operations**

```python
# Convert the temperatures_w1 from Celsius to Fahrenheit using the formula: Fahrenheit = (Celsius * 9/5) + 32

temperatures_w1_fahrenheit = (temparature_w1 * 9/5) + 32
print("Temperatures_w1_fahrenheit:", temperatures_w1_fahrenheit )
```

```python
# Find the maximum, minimum, and mean temperatures for the week

# Maximum
print("Maximum:", np.max(temperatures_w1_fahrenheit))
```

```python
# Minimum
print("Minimum:", np.min(temperatures_w1_fahrenheit))
```

```python

# Mean
print("Mean:", np.mean(temperatures_w1_fahrenheit))
```

**4. Array Slicing and Indexing**

```python
#  Extract the temperatures for the first three days of the week

temperatures_w1_first_three_days = temperatures_w1_fahrenheit[:3]
print("First three days:", temperatures_w1_first_three_days)
```

```python
# Extract the temperatures for the weekend (last two days)

temperatures_w1_weekend = temperatures_w1_fahrenheit[-2:]
print("Last two days:", temperatures_w1_weekend)
```

```python
# Extract the temperatures for the middle three days of the week

temperatures_w1_middle_three_days = temperatures_w1_fahrenheit[1:4]
print("Middle days:", temperatures_w1_middle_three_days)
```

##Pandas Series

**1. Creating Pandas Series**

```python
# Create a Pandas Series named marks
import pandas as pd

# Marks

Marks = pd.Series([ 95, 92, 89, 85, 80 ])
Custom_indices = pd.Series(['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5'])

print(Marks)
print(Custom_indices)
```

**2. Indexing and Slicing**

```python
# Use the integer index position to access the mark of the 1st rank student

print("1st Rank mark:", Marks[0])
```

```python
# Use the loc accessor to retrieve the marks of the top 3 ranks by specifying their index labels

# Recreate Marks Series with custom indices to enable .loc with string labels
Marks = pd.Series([95, 92, 89, 85, 80], index=Custom_indices.values)

print("Top 3 ranks:", Marks.loc['Rank1':'Rank3'])
```

```python
# Use the iloc accessor to retrieve the mark of the 3rd rank student

print("3rd Rank mark:", Marks.iloc[2])
```

```python
# Apply a boolean mask to filter the Series and retrieve the ranks where the marks are greater than 90

print("Greater than 90:", Marks[Marks > 90])
```

**3. Manipulating Series**

```python
# Modify the mark of the student with the 1st rank to 100

Marks['Rank1'] = 100
print("Modified Marks:", Marks)
```

```python
# Remove the entry corresponding to the last rank from the Series

Marks = Marks.drop('Rank5')
print("Modified Marks:", Marks)
```

```python
# Compute the CGPA by dividing each mark in the Series by 10

CGPA = Marks / 10
print("CGPA:", CGPA)
```

##Pandas DataFrame

**1. Creating Pandas DataFrame**

```python
# Create a Pandas DataFrame named transactions with the following data

# TransactionID: 101, 102, 103, 104, 105, 106, 107, 108, 109, 110
# ProductCategory: 'Electronics', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Furniture', 'Electronics'
# Region: 'North', 'South', 'North', 'East', 'West', 'North', 'East', 'West', 'South', 'North'
# Amount: 200, 150, 300, 450, 200, 250, 300, 180, 350, 400

import pandas as pd

data = {
    'TransactionID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'ProductCategory': ['Electronics', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Furniture', 'Electronics'],
    'Region': ['North', 'South', 'North', 'East', 'West', 'North', 'East', 'West', 'South', 'North'],
    'Amount': [200, 150, 300, 450, 200, 250, 300, 180, 350, 400]
}
```

**2. Data Exploration**

```python
# Display the transactions DataFrame and its basic information, including head, tail, shape, column names, and data types

# Display Transaction Data frame

transactions = pd.DataFrame(data)
transactions

```

```python
# Head

print("Head:", transactions.head())
transactions.head()
```

```python
# Tail

print("Tail:", transactions.tail())
transactions.tail()
```

```python
# Column Names

print("Column Names:", transactions.columns)
```

```python
# Data type

print("Data Types:", transactions.dtypes)
transactions.dtypes
```

```python
#  Display only the 'ProductCategory' and 'Amount' columns

print(transactions[['ProductCategory', 'Amount']])
transactions[['ProductCategory', 'Amount']]
```

```python
# Retrieve the last 3 columns of the DataFrame using iloc or loc

print(transactions.iloc[:,-3:])
transactions.iloc[:,-3:]
```

```python
# Filter rows where the 'Region' is 'North' and 'Amount' is greater than 200

print(transactions[(transactions['Region'] == 'North') & (transactions['Amount'] > 200)])
transactions[(transactions['Region'] == 'North') & (transactions['Amount'] > 200)]

```

```python
# Find the value counts for the 'ProductCategory' column

print(transactions['ProductCategory'].value_counts())
transactions['ProductCategory'].value_counts()
```

```python
# Find the unique values in the 'Region' column

print(transactions['Region'].unique())
```

```python
# Group by 'Region' and find the mean amount for each region

print(transactions.groupby('Region')['Amount'].mean())
transactions.groupby('Region')['Amount'].mean()
```

**3. Manipulating the DataFrame**

```python
# Modify the 'Amount' for the transaction with TransactionID 102 to 165

transactions.loc[transactions['TransactionID'] == 102, 'Amount'] = 165
print(transactions)
transactions.loc[transactions['TransactionID'] == 102]
```

```python
#  Add a new column 'Discount' by calculating 10% of the 'Amount'

transactions['Discount'] = 0.1 * transactions['Amount']
print(transactions)
transactions['Discount']
```

```python
# Remove the row with TransactionID 109

transactions = transactions[transactions['TransactionID'] != 109]
print(transactions)
transactions[transactions['TransactionID'] != 109]
```

```python
# Delete the 'Discount' column from the DataFrame

transactions = transactions.drop('Discount', axis=1, errors='ignore')
print(transactions)
transactions.drop('Discount', axis=1, errors='ignore')
```
